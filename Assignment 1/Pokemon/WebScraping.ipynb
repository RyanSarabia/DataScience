### Import `requests` library

```python
import requests

URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
```

### Load the page

```python
page=requests.get(URL)
```

### Parse HTML data

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

### Find all tables that contain Pokemon details

```python
# Get main content <div>
poke_content=soup.find(id='mw-content-text')

# Get all <table> elements
poke_tables=poke_content.find_all('table')
```

### Get list of First Generation Pokemons

```python
gen1_list=poke_tables[1]
gen2_list = poke_tables[2]
gen3_list = poke_tables[3]
gen4_list=poke_tables[4]
gen5_list = poke_tables[5]
gen6_list = poke_tables[6]
gen7_list = poke_tables[7]
gen8_list = poke_tables[8]
```

```python
info_start=3

# Let's figure out how to get each item for Bulbasaur
info_row1=gen1_list.contents[info_start]
info_row2=gen2_list.contents[info_start]
info_row3=gen3_list.contents[info_start]
info_row4=gen4_list.contents[info_start]
info_row5=gen5_list.contents[info_start]
info_row6=gen6_list.contents[info_start]
info_row7=gen7_list.contents[info_start]
info_row8=gen8_list.contents[info_start]
```
